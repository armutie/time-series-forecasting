```python
import yfinance as yf
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt
```

```python
sp500 = yf.download('^GSPC', start='2010-01-01', end='2025-02-20')
prices = sp500['Close'].values  

plt.figure(figsize=(10, 6))
plt.plot(sp500.index, prices)
plt.title('S&P 500 Closing Prices (2010–2025)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.show()

print(f"Data shape: {prices.shape}")
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Normalize the prices to [0, 1]
scaler = MinMaxScaler()
scaled_prices = scaler.fit_transform(prices.reshape(-1, 1))

# Create sequences for LSTM
look_back = 20  # Use 20 days to predict the next day
X, y = [], []
for i in range(len(scaled_prices) - look_back):
    X.append(scaled_prices[i:i + look_back])  # 20-day window
    y.append(scaled_prices[i + look_back])   # Next day’s price

X = np.array(X)  # Shape: [samples, look_back, features]
y = np.array(y)  # Shape: [samples, 1]

train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Build the model
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(look_back, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Plot training progress
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('LSTM Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)

# Inverse transform to get actual prices (not normalized)
y_pred = scaler.inverse_transform(y_pred)
y_test_actual = scaler.inverse_transform(y_test)

rmse = np.sqrt(mean_squared_error(y_test_actual, y_pred))
print(f"Root Mean Squared Error: {rmse}")

# Plot predictions vs actuals
plt.figure(figsize=(12, 6))
plt.plot(y_test_actual, label='Actual S&P 500 Prices')
plt.plot(y_pred, label='Predicted S&P 500 Prices')
plt.title('S&P 500 Price Prediction vs Actual')
plt.xlabel('Time (days in test set)')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test_actual[-100:], label='Actual')
plt.plot(y_pred[-100:], label='Predicted')
plt.title('Last 100 Days: Actual vs Predicted')
plt.legend()
plt.show()
```

```python
#----------Trying to predict all 100 future days with compunding errors----------------------
```

```python
# Start with the last known sequence from the test set
recursive_input = X_test[-1].reshape(1, look_back, 1)  

future_predictions = []

num_future_days = 100  # How many days to predict recursively

for _ in range(num_future_days):
    next_pred = model.predict(recursive_input)  
    future_predictions.append(next_pred[0, 0]) 
    
    # Shift input window: remove oldest value, add new prediction
    recursive_input = np.roll(recursive_input, shift=-1, axis=1)
    recursive_input[0, -1, 0] = next_pred  # Insert new prediction

future_predictions = np.array(future_predictions).reshape(-1, 1)
future_predictions = scaler.inverse_transform(future_predictions)

plt.figure(figsize=(12, 6))
plt.plot(y_test_actual[-100:], label='Actual (last 100 days)')
plt.plot(future_predictions, label='Recursive Forecast (next 100 days)')
plt.title('Recursive LSTM Forecast with Compounding Errors')
plt.legend()
plt.show()
```
